```python
from requests import get
from bs4 import BeautifulSoup
```

# ページ部取得

```python
header = {"User-Agent":"Mozilla/5.0"}
param = {"school_id":11,"department_id":"12","year":2025,"lang":"ja"}
r=get("https://syllabus.kosen-k.go.jp/Pages/PublicSubjects",headers=header,params=param)
```

# HTMLのパース

```python
soup = BeautifulSoup(r.text,"html.parser")
```

## 教材リスト作成

```python
t=soup.find("table",id="sytablenc").find_all("tr")
del t[0:4]
```

## 教材Listごとの処理

```python
for s in t:
    pass
    sb=s.find("div",class_="subject-item").find("a")
    name=sb.get_text()
    url=sb.get("href")
print([s.find("div",class_="subject-item").find("a").get_text() for s in t])
```
